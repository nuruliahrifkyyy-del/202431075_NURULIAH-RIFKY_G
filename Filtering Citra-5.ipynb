# Import Library

```python
#202431075_Nuruliah Rifky
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Membaca Gambar

```python
#202431075_Nuruliah Rifky
citra_m = cv2.imread("foto.jpg")

citra_m = cv2.cvtColor(citra_m, cv2.COLOR_BGR2RGB)
tinggi, lebar, channel = citra_m.shape
print(tinggi, lebar)
```

# Median Filtering

```python
#202431075_Nuruliah Rifky
img_median = citra_m.copy()
img_median_after = cv2.medianBlur(img_median,5)
```

# Mempersiapkan Mean Filtering

```python
#202431075_Nuruliah Rifky
copyCitra1 = cv2.cvtColor(citra_m, cv2.COLOR_RGB2GRAY)

copyCitra1 = copyCitra1.copy().astype(float)
m1, n1 = copyCitra1.shape
output1 = np.empty([m1,n1])
```

# Mean Filtering (Manual)

```python
#202431075_Nuruliah Rifky
for baris in range(1,m1-1):

    for kolom in range(1,n1-1):

        jumlah = (
            copyCitra1[baris-1,kolom-1] +
            copyCitra1[baris-1,kolom] +
            copyCitra1[baris-1,kolom+1] +

            copyCitra1[baris,kolom-1] +
            copyCitra1[baris,kolom] +
            copyCitra1[baris,kolom+1] +

            copyCitra1[baris+1,kolom-1] +
            copyCitra1[baris+1,kolom] +
            copyCitra1[baris+1,kolom+1]
        )

        output1[baris,kolom] = jumlah/9

hasil_mean = np.asarray(output1,dtype=np.uint8)
```

# Menampilkan Hasil

```python
#202431075_Nuruliah Rifky
fig, axs = plt.subplots(2,2, figsize=(12,8))

axs = axs.ravel()

axs[0].imshow(citra_m)
axs[0].set_title("Citra Asli")

axs[1].imshow(img_median_after)
axs[1].set_title("After filter median")

axs[2].imshow(copyCitra1, cmap="gray")
axs[2].set_title("Original Image")

axs[3].imshow(hasil_mean, cmap="gray")
axs[3].set_title("Mean Filtered Image")

plt.tight_layout()
plt.show()
```
